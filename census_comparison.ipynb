```python
import pandas as pd
```

```python
coa_2020 = pd.read_csv('Data/COA_data/Y2020.csv')
state_coa_2020 = coa_2020.groupby("STATE").sum(numeric_only=True)
state_coa_2020['NET PERM'] = state_coa_2020['TOTAL TO ZIP'] - state_coa_2020["TOTAL FROM ZIP"]
```

```python
state_coa_2020['NET PERM']
```

```python
coa_2021 = pd.read_csv('Data/COA_data/Y2021.csv')
state_coa_2021 = coa_2021.groupby("STATE").sum(numeric_only=True)
state_coa_2021['NET COA'] = state_coa_2021['TOTAL TO ZIP'] - state_coa_2021["TOTAL FROM ZIP"]
```

```python
state_coa_2021['NET COA']
```

```python
# Convert YYYYMM from string to datetime
coa_2020['DATE'] = pd.to_datetime(coa_2020['YYYYMM'],format='%Y%m')
coa_2020.set_index('DATE')
```

```python
coa_2020.set_index('DATE',inplace=True)
coa_2020.head()
```

```python
partial_coa_2020 = coa_2020.loc['2020-04-01':]
partial_coa_2020.head()
```

```python
# Convert YYYYMM from string to datetime
coa_2021['DATE'] = pd.to_datetime(coa_2021['YYYYMM'],format='%Y%m')
coa_2021.set_index('DATE',inplace=True)
```

```python
partial_coa_2021 = coa_2021.loc[:'2021-06-30']
partial_coa_2021.head()
```

```python
census_comparison = pd.concat([partial_coa_2020,partial_coa_2021])
```

```python
census_comparison
```

```python
states_census_comparison = census_comparison.groupby("STATE").sum(numeric_only=True)
```

```python
states_census_comparison["NET"] = states_census_comparison["TOTAL TO ZIP"] - states_census_comparison["TOTAL FROM ZIP"]
```
