```python
# ALL COA data is formatted the same so combine to large CSV
coa_2018 = pd.read_csv('Data/COA_data/Y2018.csv')
coa_2019 = pd.read_csv('Data/COA_data/Y2019.csv')
coa_2020 = pd.read_csv('Data/COA_data/Y2020.csv')
coa_2021 = pd.read_csv('Data/COA_data/Y2021.csv')
coa_2022 = pd.read_csv('Data/COA_data/Y2022.csv')
coa_2023 = pd.read_csv('Data/COA_data/Y2023.csv')

coa_data = [coa_2018, coa_2019, coa_2020, coa_2021, coa_2022, coa_2023]

coa_final = pd.concat(coa_data)
```

```python
coa_final.tail()
```

```python
# Convert YYYYMM from string to datetime
coa_final['YYYYMM'] = pd.to_datetime(coa_final['YYYYMM'],format='%Y%m')
```

```python
coa_final.dtypes
```

```python
coa_final['NET CHANGE'] = coa_final['TOTAL PERM.1'] - coa_final['TOTAL PERM']
```

```python
coa_export = coa_final[['YYYYMM','STATE','NET CHANGE']].groupby(['STATE','YYYYMM']).sum()
```

```python
coa_export.to_csv('Data/COA_data/data_of_interest.csv')
```

```python
net_change_df = pd.read_csv('Data/COA_data/data_of_interest.csv')
net_change_df
```

```python
def time_period_flag(x):
    if x > '2022-03-31':
        return '3'
    elif x > '2021-03-31':
        return '2'
    elif x > '2020-03-31':
        return '1'
    elif x > '2019-03-31':
        return '0'
    else:
        return
```

```python
net_change_df['STAGE'] = net_change_df['YYYYMM'].apply(time_period_flag)
```

```python
net_change_df
```

```python
average_change_per_month = net_change_df.groupby(['STAGE','STATE']).mean()
```

```python
new_df = pd.DataFrame({'Pre-Lockdown (PPL/MON)':average_change_per_month.loc[('0'),'NET CHANGE'],
                       '1st Year post-lockdown (PPL/MON)':average_change_per_month.loc[('1'),'NET CHANGE'],
                       '2nd Year post-lockdown (PPL/MON)':average_change_per_month.loc[('2'),'NET CHANGE'],
                       '3rd Year post-lockdown (PPL/MON)':average_change_per_month.loc[('3'),'NET CHANGE']})
```

```python
new_df
```

```python
new_df.to_csv('Data/COA_data/4_stages_of_migration.csv')
```
